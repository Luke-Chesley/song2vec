```python
from bc import bit_cypher,create_tokenizer
import numpy as np
import torch
from collections import Counter
```

```python
n = 9
u, v,vocab_size = bit_cypher(n)

vocab_size
```

```python
tokenizer = create_tokenizer('ts.txt')
```

```python
text = [x.strip() for x in open('ts.txt').readlines()]
text = [x for x in text if x != '']
```

```python
c = Counter([y for x in [tokenizer.encode(text[i]).ids for i in range(len(text))] for y in x])

top_n_counts = c.most_common(vocab_size) # cutoff for testing [token id, n_occurrences]

ids, counts = zip(*top_n_counts)

tokens_in_corpus = sum(counts)

f_i = [x/tokens_in_corpus for x in counts]

y_i = u

B_i = torch.tensor([x/(x+1) for x in f_i])

sigma = [(1-(x / tokens_in_corpus))/(vocab_size-1) for x in f_i]

#sigma = torch.tensor(sigma)
```

```python
v_i = torch.zeros_like(v)
for i in range(len(y_i)):
    v_i[i] = y_i[i] * B_i[i] + ((1-B_i[i]) * sigma)
```
